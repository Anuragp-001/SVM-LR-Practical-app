## Support vector Machine

```python
import numpy as nb
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
import joblib                                                                                           #To store the model
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
pd.DataFrame(data.data , columns=data.feature_names)
```

```python
df = pd.DataFrame(data.data , columns=data.feature_names)
```

```python
data.target               # species 0--> setosa , 1--> versicolor , 2--> varginica 
```

```python
df['target'] = data.target
```

```python
df
```

```python
data.target_names
```

```python
df["Target_names"] = df['target'].map({i:name for i , name in enumerate(data.target_names)})                 # 0 , 1 ,2 ko uske target name se add kiya 
```

```python
df
```

```python
df.head()
```

## Lets us build the binary class by neglecting the one of the class 2 or else 

```python
binary_df = df[df['target'] != 2]
```

```python
binary_df
```

```python
x_binary = binary_df[['sepal length (cm)' ,'sepal width (cm)'	, 'petal length (cm)',	'petal width (cm)']]
```

```python
x_binary
```

```python
y_binary = binary_df['target']
```

```python
y_binary
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_binary , y_binary , test_size=0.2)
```

```python
x_train.describe()
```

```python
#Lets us standardise the data 
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
svm_binary = SVC(kernel='linear',C=1.0,probability=True)        
```

```python
svm_binary.fit(x_train,y_train)
```

```python
svm_binary.predict(x_test)
```

```python
print("binary class classification reprt:\n" , classification_report(y_test,svm_binary.predict(x_test)))
```

```python
confusion_matrix(y_test,svm_binary.predict(x_test))
```

## Now we have to save the model in joblib 

```python
joblib.dump(svm_binary , 'svm_binary.pkl')         # file name --> 'svm_binary.pkl'
```

## Lets do classification for multi class classification 0 , 1 , 2

```python
x_multi = df[['sepal length (cm)' ,'sepal width (cm)'	, 'petal length (cm)',	'petal width (cm)']]
```

```python
x_multi
```

```python
y_multi = df['target']
```

```python
y_multi
```

```python
x_train_multi , x_test_multi , y_train_multi , y_test_multi = train_test_split(x_multi , y_multi , test_size=0.2)
```

```python
svm_multi = SVC(kernel='linear')
```

```python
svm_multi.fit(x_train_multi , y_train_multi)
```

```python
svm_multi.predict(x_test_multi)
```

```python
print("report for multi class svc \n" , classification_report(y_test_multi , svm_multi.predict(x_test_multi)))
```

```python
confusion_matrix(y_test_multi , svm_multi.predict(x_test_multi))
```

```python
joblib.dump(svm_multi , 'svm_multi.pkl')
joblib.dump(scaler , 'scaler.pkl')
```

# Logistic regression for multi class model 

```python
from sklearn.multiclass import OneVsRestClassifier
```

### # Note --> here we do not use syntax like {LogisticRegression(multi_class='ovr',solver='lbfgs',max_iter=2000)} instead we use  OneVsRestClassifier(LogisticRegression(solver='lbfgs',max_iter=2000)) coz new python version have The default behavior in modern scikit-learn which automatically handles multiclass problems

```python
logistic_ovr = OneVsRestClassifier(LogisticRegression(solver='lbfgs',max_iter=2000))    
```

```python
logistic_ovr.fit(x_train_multi , y_train_multi)
```

```python
logistic_ovr.predict(x_test_multi)
```

```python
print("Classificatiob report for OVR logistics \n" , classification_report(y_test_multi ,logistic_ovr.predict(x_test_multi) ))
```

```python
confusion_matrix(y_test_multi ,logistic_ovr.predict(x_test_multi))
```

```python
joblib.dump(logistic_ovr , "logistic.pkl")
```
